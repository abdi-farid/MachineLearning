# **Pyplot**

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. interpolation

```python
X = np.linspace(1, 10, 10)
Y = X **2
plt.scatter(X, Y)
```

```python
from scipy.interpolate import interp1d
```

```python
# fonction d interpolation
f = interp1d(X, Y, kind='linear')

new_X = np.linspace(1, 10, 30)
new_Y = f(new_X)

plt.scatter(X, Y)
plt.scatter(new_X, new_Y)
```

# Minimize

```python
x = np.linspace(0,2, 100)
y = 1/3*x**3 - 3/5 * x **2 + 2+ np.random.randn(x.shape[0])/20
plt.scatter(x,y)
```

### fonction modele 

```python
def f(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d
```

```python
from scipy import optimize
```

```python
optimize.curve_fit(f, x, y)
```

```python
params, params_cov = optimize.curve_fit(f, x , y)
params
```

```python
plt.scatter(x,y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]), c='red')
```

# transformation de fourier
### analyse des **frquence** prsentes dans un signal *priodique*
### en rsultat : **'Un spectre'**

## gnrer un signal alatoire

```python
x = np.linspace(0, 30, 1000)
y = 3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x)
plt.plot(x,y)
```

### import de module fftpack

```python
from scipy import fftpack
```

```python
fourier = fftpack.fft(y)
frequences = fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences), np.abs(fourier))
```
